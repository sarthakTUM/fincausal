```python
import pandas as pd
import numpy as np
from pathlib import Path
```

```python
train_path = Path('/media/sarthak/HDD/data_science/fnp_resources/data/task1/all_combined/train.csv')
dev_path = Path('/media/sarthak/HDD/data_science/fnp_resources/data/task1/all_combined/dev.csv')
test_path = Path('/media/sarthak/HDD/data_science/fnp_resources/data/task1/all_combined/test.csv')
```

```python
train_df = pd.read_csv(train_path)
dev_df = pd.read_csv(dev_path)
test_df = pd.read_csv(test_path)
```

```python
train_df.head()
```

```python
train_dev_test_df = pd.concat([train_df, dev_df, test_df])
train_dev_test_df.head()
```

```python
train_dev_test_df = train_dev_test_df.sample(n=len(train_dev_test_df))
train_dev_test_df.head()
```

```python
train_dev_test_df.to_csv(Path('/media/sarthak/HDD/data_science/fnp_resources/data/task1/all_combined/train_dev_test.csv'), index=False)
```

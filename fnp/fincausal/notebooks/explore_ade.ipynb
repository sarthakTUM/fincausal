```python
import pandas as pd
import numpy as np
import os
from pathlib import Path
```

```python
drug_ae_path = Path('/media/sarthak/HDD/data_science/fnp_resources/data/AdverseDrugReaction-master/ADE-Corpus-V2/DRUG-AE.rel')
ade_neg_path = Path('/media/sarthak/HDD/data_science/fnp_resources/data/AdverseDrugReaction-master/ADE-Corpus-V2/ADE-NEG.txt')
```

```python
def read_ae(path):
    records = []
    with open(path, 'r') as f:
        for line in f.readlines():
            records.append(line.split('|')[1])
    return records
```

```python
def read_ade_neg(path):
    records = []
    with open(path, 'r') as f:
        for line in f.readlines():
            records.append(line.split('NEG ')[1].split('\n')[0])
    return records
```

```python
drug_ae_records = list(set(read_ae(drug_ae_path)))
```

```python
ade_neg_records = list(set(read_ade_neg(ade_neg_path)))
```

```python
drug_ae_df = pd.DataFrame({
    'sentences': drug_ae_records,
    'causal': 1
})
```

```python
neg_ae_df = pd.DataFrame({
    'sentences': ade_neg_records,
    'causal': 0
})
```

```python
all_df = pd.concat([drug_ae_df, neg_ae_df], ignore_index=True)
all_df.shape
```

```python
all_df = all_df.sample(len(all_df))
```
